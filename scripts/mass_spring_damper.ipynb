```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Variables
data_dir = Path(".") / ".." / "data" / "mass_spring_damper"
```

```python
# Read csv data
euler_forward_df = pd.read_csv(data_dir / "euler_forward.csv")
runge_kutta_4_df = pd.read_csv(data_dir / "runge_kutta_4.csv")
```

```python
# Plot
plt.figure()
plt.plot(euler_forward_df["t"], euler_forward_df["y0"], label="Euler Forward")
plt.plot(runge_kutta_4_df["t"], runge_kutta_4_df["y0"], label="Runge Kutta 4")
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")
plt.legend()
plt.show()
```
